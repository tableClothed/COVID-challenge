```python
import nltk
import pandas as pd
import os
import pycountry
import numpy as np
import spacy
import warnings; warnings.simplefilter('ignore')
```

```python
nlp = spacy.load('en_core_web_lg')

DIR = "csvfiles/tasks/"
```

```python
def find_similarity(df, tags):
    task = []
    for i, row in df.iterrows():
        try:
            title = row[1].lower()
            title = nlp(title)
            result = title.similarity(tags)
            task.append([i, result])
        except Exception as e: print(e)
            
    sorted_results = sorted(task, key=lambda x: x[1], reverse=True)[:10]
    ids = [i for i, r in sorted_results]
    dataframe = df.loc[ids]
    dataframe = dataframe.reset_index(drop=True)
    prob = pd.DataFrame({"similarity": [r for i, r in sorted_results]})
    dataframe = dataframe.join(prob)     
        
    return dataframe
```

```python
def find_final_dataframe(task_num, tags):
    df = pd.read_csv(DIR + "task" + str(task_num) + ".csv")
    datas = []
    
    for tag in tags:
        paragraphs, par_id = [], []

        for i, row in df.iterrows():
            sentences = row[2].split('.')
            for sentence in sentences:
                sentence_changed = (sentence.replace('.', '').replace(' ', '').replace(';', '')
                                    .replace(':', '').replace(',', '').replace('\n', '').replace("'", '')
                                    .replace('"', '').lower())
                if len(sentence_changed) < 40:
                    continue
                if sentence_changed not in paragraphs:
                    sentence = sentence.replace('\n', '')
                    paragraphs.append(sentence_changed)
                    par_id.append([i, sentence])
        new_df = pd.DataFrame(par_id)
        new_df.columns = ['id', 'text']
        final_df = find_similarity(new_df, tag)
        final_df = final_df.join(df[['title']], on='id')
        datas.append(final_df)
    return datas
```

```python
def read_result(task):
    for i, r in task.iterrows():
        print('------------------------')
        print(r[0])
        print('------------------------')
        print(r[1])
        print('------------------------')
        print('\n\n')
```
