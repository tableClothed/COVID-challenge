```python
import pandas as pd
import numpy as np
import os
import spacy
import warnings; warnings.simplefilter('ignore')
```

```python
nlp = spacy.load('en_core_web_lg')
DIR = "csvfiles/tasks/"

for i in os.listdir(DIR):
    print(i)
```

```python
def show_relevant_articles(task_num, tags):
    df = pd.read_csv(DIR + f"task{task_num}.csv")
    inf = []
    sentences = []
    
    for i, row in df.iterrows():
        if len(row[2]) <= 50:
            break
        akapity = row[2].split('\n')
        a = row[1].split('\n')
        akapity.append(a)
        try:
            for akapit in akapity:
                if len(akapit) <= 500:
                    break
                tag = nlp(tags)
                ak = nlp(akapit.lower())
                similarity = tag.similarity(ak)
                inf.append([akapit, similarity])
        except: pass
        sorted_results = sorted(inf, key=lambda x: x[1], reverse=True)[:2]
        print("----------------------------------------")
        print(row[0].upper())
        print("----------------------------------------" + "\n")
        for j in sorted_results:
            print(j[0] + "\n\n")
```

```python
def find_duplic(infos):
    df = pd.DataFrame(infos)
    df.head()
```
