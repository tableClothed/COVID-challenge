```python
import pandas as pd
import os

DIR = "csvfiles/relevant/"
```

```python
BIG_DF = pd.read_csv("csvfiles/metadata.csv").drop_duplicates()
```

```python
BIG_DF = BIG_DF[["title", "authors", "publish_time", "url"]]
BIG_DF
```

```python
for csvfile in os.listdir(DIR):
    titles, articles = [], []
    df = pd.read_csv(DIR + csvfile)
    for j, row in df.iterrows():
        title = row['title']
        text = row['text']
        if title not in titles:
            if "The incubation period is defined as the time between infection and onset of symptoms" not in text:
                titles.append(title)
                articles.append(text)
        else:
            i = titles.index(title)
            articles[i] += "\n" + text
    new_df = pd.DataFrame({"title": titles, "text": articles})
    new_df = pd.merge(new_df, BIG_DF, on="title")
    new_df = new_df.sort_values('publish_time', ascending=False)
    new_df.to_csv(DIR + csvfile, index=False)
```
